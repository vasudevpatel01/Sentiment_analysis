```python
from googletrans import Translator
from nltk.sentiment.vader import SentimentIntensityAnalyzer

translator = Translator()
sia = SentimentIntensityAnalyzer()

# Assuming your DataFrame from NewsAPI has a 'language' column
# Translate 'title' column
title_translated = []
for idx, row in df.iterrows():
    text = str(row['title'])
    
    
    text_en = translator.translate(text, dest='en').text
   
    title_translated.append(text_en)

df['title_en'] = title_translated

# Translate 'description' column
description_translated = []
for idx, row in df.iterrows():
    text = str(row['description'])
    text_en = translator.translate(text, dest='en').text

    description_translated.append(text_en)

df['description_en'] = description_translated

# Apply VADER sentiment analysis on translated text
df['title_sentiment'] = df['title_en'].apply(lambda x: sia.polarity_scores(x)['compound'])
df['description_sentiment'] = df['description_en'].apply(lambda x: sia.polarity_scores(x)['compound'])
df['overall_sentiment'] = df[['title_sentiment', 'description_sentiment']].mean(axis=1)
 
```
